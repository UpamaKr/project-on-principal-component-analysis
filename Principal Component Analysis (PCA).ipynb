# Principal Component Analysis

Principal Component Analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables into a set of values of linearly uncorrelated variables called principal components. This transformation is defined in such a way that the first principal component has the largest possible variance, and each succeeding component in turn has the highest variance possible under the constraint that it is orthogonal to the preceding components. The resulting vectors are an uncorrelated orthogonal basis set. PCA is sensitive to the relative scaling of the original variables.

Steps to perform Principal Component Analysis (PCA) in Python without the code:

1. Import the necessary libraries: Libraries such as sklearn for the PCA algorithm and pandas or numpy for data manipulation are required.

2. Load the data: The data should be numeric and stored as NumPy arrays or pandas DataFrame.

3. Standardize the Data: Since PCA is affected by scale, the features in the data need to be scaled before applying PCA. StandardScaler is used to standardize the features to have mean=0 and variance=1.

4. Create the PCA model: The PCA class in sklearn is used to perform PCA.

5. Fit the model: The fit method of the PCA class is used to fit the model to the data.

6. Transform the data: The transform method of the PCA class is used to transform the data to its first 2 principal components.

This is a simple example. In a real-world scenario, the data would also need to be split into a training set and a test set, the data would need to be standardized, and the parameters of the PCA would need to be tuned, among other things.

## PCA Visualization


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

to  specify how many components we want to keep when creating the PCA object.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

transform this data to its first 2 principal components.

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], edgecolors='black',linewidths=0.5)
# plot all the rows from 0 against all the rows from 1
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma', edgecolors='black', linewidths=0.5)
# plot all the rows from 0 against all the rows from 1
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Interpreting the components
Unfortunately, with this great power of dimensionality reduction, comes the cost of being able to easily understand what these components represent.

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object:

```python
pca.components_
```

In this numpy matrix array, each row represents a principal component, and each column relates back to the original features. we can visualize this relationship with a heatmap:

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself.

higher the number is more corelated to a specific fiture in the columns.
